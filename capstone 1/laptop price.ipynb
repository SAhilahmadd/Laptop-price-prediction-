## Outline Of Machine learning Project

Predicting the price of a laptop is a supervised learning problem in the field of machine learning. We can use various algorithms like linear regression, decision trees, random forest, etc. to build a model that can predict the price of a laptop given certain features like the brand, processor, graphics card, storage, RAM, etc.


1. Collect Data = We need to gather data on laptops and their prices. We can use websites such as Amazon, Best Buy, etc.to get                  this information. The data should include features such as brand, processor, graphics card,storage,RAM,etc.and                   the corresponding price of each laptop Data was collected for this project from Analytics Vidhya .



2. Preprocess Data = Once we have collected the data, you need to clean and preprocess it. This may include handling missing                        values, removing duplicates, and normalizing the data .



3. Split Data = Divide the preprocessed data into training and testing sets. You can use the training set to train your model                   and the testing set to evaluate its performance.



4. Choose an Algorithm = Choose a machine learning algorithm that you want to use for the prediction. You can start with simple                          algorithms like linear regression or decision trees and then move on to more complex algorithms like                            Adda Boost or gradient boosting.



5. Train the Model = Train the chosen algorithm on the training set. This will allow the model to learn the relationshipbetween                     the features and the price.



6. Evaluate the Model = Evaluate the performance of the model on the testing set. You can use metrics such as mean absolute                             error, root mean squared error, etc. to determine the accuracy of the model.



7. Fine-Tune the Model = If the performance of the model is not satisfactory, you can fine-tune it by changing the algorithm,                            adding/removing features, or using regularization techniques.



8. Make Predictions = Finally, you can use the trained and fine-tuned model to make predictions on new data.


This is a basic outline of how we can build a laptop price prediction model using Python. 

## Variables Used in Project


1. Company = This refers to the name of the company that manufactures the laptop.


2. TypeName = This refers to the name or model of the laptop.



3. Inches = This refers to the size of the laptop screen in inches.



4. ScreenResolution = This refers to the resolution of the laptop screen, typically represented as the number of pixels in width                     x height (e.g. 1920x1080).



5. Cpu = This refers to the central processing unit (CPU) in the laptop, which is the main component that performs computations.



6. Ram = This refers to the random access memory (RAM) in the laptop, which is a type of memory that is used to temporarily             store data while the laptop is running.



7. Memory = This refers to the amount of storage space available in the laptop, typically represented in gigabytes (GB) or                 terabytes (TB).



8. Gpu = This refers to the graphics processing unit (GPU) in the laptop, which is responsible for rendering images and videos.



9. OpSys =  This refers to the operating system (OS) installed on the laptop, such as Windows, MacOS, or Linux.



10. Weight = This refers to the weight of the laptop, typically represented in pounds (lbs) or kilograms (kg).



11. Price = This refers to the cost of the laptop, typically represented in a currency like US dollars (USD), British pounds                (GBP), or euros (EUR).

## Problem Statement for Laptop Price Prediction 

The problem statement of the above laptop price prediction project would be to predict the price of a laptop given its various features such as brand, screen size, screen resolution, processor, RAM, storage, graphics card, operating system, weight, etc.

The goal of the project is to build a machine learning model that can learn the relationship between the laptop's features and its price, and use that information to predict the price of a new laptop given its features. This model can be useful for consumers who are looking to buy a laptop, as well as for retailers who want to set competitive prices for their products.

In order to achieve this goal, you need to collect and preprocess a large dataset of laptops and their prices, choose a suitable machine learning algorithm, train the model on the data, evaluate its performance, and fine-tune the model as needed. The final model should be able to make accurate price predictions for new laptops with a high degree of confidence.

We have lots of hidden information which in the memory variable like HDD, SDD etc. We are going to extract all the information from it and made new different variables.Also we vanishes or reassembeled the extra information which is not usefull for us in model development like special symbol, extra space ,str_value etc.

```python
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')
new = df["Memory"].str.split("+", n = 1, expand = True)

df["first"]= new[0]
df["first"]=df["first"].str.strip()

df["second"]= new[1]

df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['first'] = df['first'].str.replace(r'\D', '')

df["second"].fillna("0", inplace = True)

df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['second'] = df['second'].str.replace(r'\D', '')

df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)

df["HDD"]=(df["first"]*df["Layer1HDD"]+df["second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["first"]*df["Layer1Flash_Storage"]+df["second"]*df["Layer2Flash_Storage"])

df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

## Model devlopment

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

In above column we are going to split data into x and y Crucial step in model development

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

## Linear Regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

## Ridge

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = Ridge(alpha=10)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

## Lasso

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = Lasso(alpha=0.001)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

## KNN

```python
# knn
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = KNeighborsRegressor(n_neighbors=3)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

## Decission Tree

```python
# decission tree
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = DecisionTreeRegressor(max_depth=8)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

## SVM

```python
# svm
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = SVR(kernel='rbf',C=10000,epsilon=0.1)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

## Random Forest

```python
# random forest
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

## Adda boost

```python
# a
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = AdaBoostRegressor(n_estimators=15,learning_rate=1.0)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

## Gradient

```python
# a
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = GradientBoostingRegressor(n_estimators=500)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

## xgboost

```python
# xgboost

step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')

step2 = XGBRegressor(n_estimators=45,max_depth=5,learning_rate=0.5)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
# stacking
from sklearn.ensemble import VotingRegressor,StackingRegressor

step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,1,7,10,11])
],remainder='passthrough')


estimators = [
    ('rf', RandomForestRegressor(n_estimators=350,random_state=3,max_samples=0.5,max_features=0.75,max_depth=15)),
    ('gbdt',GradientBoostingRegressor(n_estimators=100,max_features=0.5)),
    ('xgb', XGBRegressor(n_estimators=25,learning_rate=0.3,max_depth=5))
]

step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
